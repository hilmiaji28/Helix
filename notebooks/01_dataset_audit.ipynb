```python
# import library
from pathlib import Path

import warnings

warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns

from IPython.display import display

plt.style.use("ggplot")

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 200)
```

```python
# load dataset
PROJECT_ROOT = Path("..")

DATA_PATH = PROJECT_ROOT / "data" / "raw" / "ecommerce_customer_behavior_dataset_v2.csv"

df = pd.read_csv(DATA_PATH)

print(df.shape)

df.head()
```

```python
# dataset overview
display(df.head())

display(df.tail())

display(df.sample(5, random_state=42))
```

```python
# dataset dimension
rows, cols = df.shape

summary = pd.DataFrame({"Metric": ["Rows", "Columns"], "Value": [rows, cols]})

display(summary)
```

```python
# dataset information
df.info(memory_usage="deep")
```

```python
# memory usage
memory = df.memory_usage(deep=True).to_frame("Memory (Bytes)")

memory["Memory (MB)"] = memory["Memory (Bytes)"] / 1024 / 1024

display(memory)

print(f"Total Memory : {memory['Memory (MB)'].sum():.2f} MB")
```

```python
# missing value audit
missing = pd.DataFrame(
    {"Missing": df.isna().sum(), "Percentage": (df.isna().mean() * 100).round(2)}
)

missing = missing.sort_values("Missing", ascending=False)

display(missing)
```

```python
# duplicate audit
duplicate = df.duplicated().sum()

print(f"Duplicate Rows : {duplicate}")
```

```python
# unique value audit
unique = pd.DataFrame({"Unique": df.nunique()})

display(unique)
```

```python
# cardinality audit
cardinality = df.nunique().sort_values(ascending=False).to_frame("Unique Values")

display(cardinality)
```

```python
# numerical feature
num_cols = df.select_dtypes(include=np.number).columns

display(df[num_cols].describe().T)
```

```python
# categorical feature
cat_cols = df.select_dtypes(include="object").columns

for col in cat_cols:
    print("=" * 80)

    print(col)

    display(df[col].value_counts().to_frame())
```

```python
# datetime
df["Date"] = pd.to_datetime(df["Date"])

print(df["Date"].min())

print(df["Date"].max())
```

```python
# target audit
target = "Total_Amount"

display(df[target].describe())

print("Skewness:", df[target].skew())

print("Kurtosis:", df[target].kurtosis())
```

```python
# initial leakage detection
check = df["Unit_Price"] * df["Quantity"] - df["Discount_Amount"]

difference = (check - df["Total_Amount"]).abs().max()

print(f"Maximum Difference : {difference:.10f}")
```

## Target Variable Assessment

The prediction target is:

**Total_Amount**

Initial statistical analysis indicates that:

- The distribution is highly right-skewed.
- The mean is considerably higher than the median, suggesting the presence of high-value transactions.
- Outlier analysis will be performed in the next notebook to determine whether transformation or robust modeling techniques are required.

---

## Potential Target Leakage

A preliminary investigation suggests that:

The computed values are nearly identical to the target, indicating that these three variables may directly determine the prediction target.

If confirmed, using these variables in the predictive model would produce artificially high performance while providing limited business value.

Therefore, two modeling scenarios are proposed:

### Scenario A — Business Revenue Prediction

Predict revenue using customer characteristics and behavioral features only.

Candidate features include:

- Age
- Gender
- City
- Product_Category
- Payment_Method
- Device_Type
- Session_Duration_Minutes
- Pages_Viewed
- Is_Returning_Customer
- Delivery_Time_Days
- Customer_Rating
- Engineered datetime features

This scenario reflects a realistic business prediction problem.

---

### Scenario B — Transaction Amount Estimation

Include:

- Unit_Price
- Quantity
- Discount_Amount

This scenario estimates transaction totals once purchase information is already available.

Although prediction accuracy is expected to be substantially higher, the resulting model is less valuable for early-stage business decision making.
